```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("/content/front_face.jpg",0)
he_img=cv2.equalizeHist(img)

plt.figure(figsize=(12,5))
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Histogram Equalized")
plt.imshow(he_img)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Histogram Equalization")
plt.hist(he_img.ravel(),bins=256)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Read original image in COLOR
img_color = cv2.imread("/content/front_face.jpg")        # BGR format
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)   # convert to RGB for matplotlib

# Convert to grayscale for histogram equalization
img_gray = cv2.cvtColor(img_color, cv2.COLOR_RGB2GRAY)

# Apply histogram equalization
he_img = cv2.equalizeHist(img_gray)

plt.figure(figsize=(12,5))

# Show original COLOR image
plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(img_color)
plt.axis("off")

# Show equalized grayscale image
plt.subplot(1,3,2)
plt.title("Histogram Equalized")
plt.imshow(he_img, cmap='gray')
plt.axis("off")

# Show histogram
plt.subplot(1,3,3)
plt.title("Histogram")
plt.hist(he_img.ravel(), bins=256)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load original image in grayscale
img = cv2.imread("/content/elephant.jpg", 0)

clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(img)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)   # TRUE grayscale
plt.axis("off")

plt.subplot(1,2,2)
plt.title("CLAHE Enhanced")
plt.imshow(clahe_img, cmap='gray', vmin=0, vmax=255)  # TRUE grayscale
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
img =cv2.imread("/content/zoo.jpg")
img_ycrcb=cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
Y,Cr,Cb =cv2.split(img_ycrcb)
clahe=cv2.createCLAHE(clipLimit=2.5,tileGridSize=(8,8))
Y_clahe=clahe.apply(Y)
img_clahe=cv2.merge((Y_clahe,Cr,Cb))
img_clahe=cv2.cvtColor(img_clahe,cv2.COLOR_YCrCb2BGR)
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.subplot(1,2,2)
plt.title("COLOR ENHANCED")
plt.imshow(cv2.cvtColor(img_clahe,cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```
